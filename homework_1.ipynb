```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
df = pd.read_csv("bank-additional-full.csv",sep=';')

df.shape
```

### df.info()

Purpose: Provides a concise summary of the DataFrame.
What it shows:

* Number of rows and columns.
* Column names and their data types (int64, float64, object, etc.).
* Non-null count for each column (helps identify missing values).
* Memory usage of the DataFrame.

Why it’s useful: Quickly checks the structure of the dataset and detects columns with missing data or incorrect types.


### df.describe()

Purpose: Generates descriptive statistics for numeric columns by default.
What it shows:

* count: Number of non-null values.
* mean: Average value.
* std: Standard deviation.
* min, max: Minimum and maximum values.
* 25%, 50%, 75%: Quartiles (percentiles).

Why it’s useful: Gives an overview of the distribution and scale of numeric features, helping spot outliers or unusual ranges.

```python
# 1. Histogram for numeric column (age)
plt.figure(figsize=(8, 5))  # Set the figure size to 8x5 inches for better readability

# Create a histogram using Seaborn:
# - df['age']: the column to plot
# - bins=30: divide the age range into 30 intervals
# - kde=True: overlay a Kernel Density Estimate curve for smooth distribution
# - color='skyblue': set the bar color to sky blue
sns.histplot(df['age'], bins=30, kde=True, color='skyblue')

# Add a title and axis labels for clarity
plt.title('Age Distribution')   # Title of the plot
plt.xlabel('Age')               # Label for the X-axis
plt.ylabel('Frequency')         # Label for the Y-axis

plt.show()
```

```python
# Get counts for each category
counts = df['job'].value_counts()
categories = counts.index
values = counts.values

# Normalize values for colormap
norm = plt.Normalize(values.min(), values.max())
colors = [plt.cm.viridis(norm(v)) for v in values]  # Convert to list of colors

# Plot bars with hue workaround
plt.figure(figsize=(10, 6))
sns.barplot(
    y=categories,
    x=values,
    hue=categories,       # Assign hue to same variable to allow palette
    palette=colors,       # Pass list of colors
    legend=False          # Hide redundant legend
)
plt.title('Job Distribution')
plt.xlabel('Count')
plt.ylabel('Job');
```

```python
# 3. Target variable distribution (y)
# in column 'y' - has the client subscribed a term deposit?

plt.figure(figsize=(6, 4))
sns.countplot(x='y', data=df, palette='pastel', hue='y')
plt.title('Has the client subscribed a term deposit')
plt.xlabel('Subscription')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Correlation heatmap for numeric features
plt.figure(figsize=(12, 8))

# Select only numeric columns for correlation
corr = df.select_dtypes(include=['int64', 'float64']).corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Numeric Features)');
```

```python
# Detect outliers in 'age' using IQR
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

outliers_age = df[(df['age'] < (Q1 - 1.5 * IQR)) | (df['age'] > (Q3 + 1.5 * IQR))]
print("Number of age outliers:", len(outliers_age));

#len(outliers_age) shows how many rows have an age considered an outlier
#It identifies customers whose age is unusually low or high compared to the majority.
```

```python

from sklearn.ensemble import IsolationForest

numeric_cols = df.select_dtypes(include=['int64', 'float64'])
model = IsolationForest(contamination=0.01, random_state=42)
df['anomaly_score'] = model.fit_predict(numeric_cols)
print(df['anomaly_score'].value_counts())  # -1 = anomaly


# Isolation Forest detects multivariate anomalies, i.e., unusual combinations
# of numeric features (not just single-column outliers).
# These anomalies could represent customers with very atypical profiles
# (e.g., extremely high duration and very low campaign count).
```
